```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


file_path = r'C:\Users\vikto\Documents\Skole\Datadrevne_systemer\datasets\adult\adult.data'
df = pd.read_csv(file_path, header=None)

df.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num',
              'marital-status', 'occupation', 'relationship', 'race', 'sex',
              'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']

```

```python
df.info
```

```python
sns.pairplot(df, hue="income", markers=["o", "s", "D"])
#plt.show()
```

### Distrobution of dataset: 

```python
# Count the occurrences of each income category
income_counts = df['income'].value_counts()

# Create a bar plot
sns.barplot(x=income_counts.index, y=income_counts.values)

# Add labels and title
plt.xlabel('Income Category')
plt.ylabel('Count')
plt.title('Income Category Counts')

# Show the plot
plt.show()
```

## AGE distrobution

```python
# Create a histogram of age with hue for income
sns.histplot(data=df, x='age', hue='income', multiple='stack', palette='pastel')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution by Income')

# Show the plot
plt.show()
```
